```python
# Copy your original DataFrame
df_encoded = df.copy()

# Define the custom mapping
leukemia_risk_mapping = {
    'Low': 0,
    'Moderate': 1,
    'High': 2,
    'Confirmed': 3
}

df_encoded['leukemia_risk'] = df_encoded['leukemia_risk'].map(leukemia_risk_mapping)
```

```python
# Define the custom mapping for gender
gender_mapping = {
    'Male': 1,
    'Female': 0,
}

df_encoded['gender'] = df_encoded['gender'].map(gender_mapping)
```

```python
df_encoded.head(10)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
df_encoded_train, df_encoded_test = train_test_split(df_encoded, test_size=0.3)
```

```python
df_encoded_train.shape
```

```python
X_train = df_encoded_train.drop(columns=["leukemia_risk"]).values
```

```python
X_train.shape
```

```python
Y_train = df_encoded_train["leukemia_risk"].values
```

```python
Y_train.shape
```

```python
X_test = df_encoded_test.drop(columns=["leukemia_risk"]).values
```

```python
Y_test = df_encoded_test['leukemia_risk'].values
```

Importing the CatboostClassifier Model again

```python
# Since the CatBoostClassifier score was 97 I decided to use it as the baseline model.

from catboost import CatBoostClassifier
model = CatBoostClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

Cross-Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model, X_train, Y_train, cv=5)
print(f"Cross-validation accuracy: {np.mean(scores):.2f}")
```
